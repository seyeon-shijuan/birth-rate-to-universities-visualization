```python
df2=df2.replace("-",0)
```

```python
dfc=df2.columns
```

```python
#글자 색 코드 랜덤 추출
import random
colran=range(0,256**3)
cols=random.sample(colran,k=len(dfc))
for i in range(len(cols)):
  t=format(cols[i],"06x")
  f='#'+t
  cols[i]=f
print(cols)
```
