```python
#지역별 출생아수 데이터 불러오기
df2=pd.read_csv("https://raw.githubusercontent.com/seyeon-shijuan/birth-rate-to-universities-visualization/master/data/newborn_by_city.csv",encoding="cp949")

df2
```


 그래프6) 지역별 1980년~2070년 신생아 수 꺾은선 그래프

 


```python
# 0 대신 들어간 하이픈(-)처리
df2=df2.replace("-",0)
```

```python
#칼럼명 추출(시 도 이름 추출)
dfc=df2.columns
```

```python
#글자 색 코드 랜덤 추출
import random
colran=range(0,256**3)
cols=random.sample(colran,k=len(dfc))
for i in range(len(cols)):
  t=format(cols[i],"06x")
  f='#'+t
  cols[i]=f
print(cols)
```

```python
#그래프 생성 
fig = plt.figure(figsize=(20,12))
ax1=fig.add_subplot(111)
i= 0
for char in dfc:
  if i > 2: 
    ax1.plot(df2["시점"],df2[char],label=char,marker='o',color=cols[i])
  i=i+1
plt.legend()
ax1.set_xlabel("연도")
ax1.set_ylabel("출생아수")
```

```python
plt.savefig('newborn_by_city.png')
```
